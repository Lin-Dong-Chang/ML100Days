#### 1. 畫出cos圖檔並儲存

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,360)
y = np.sin(x*np.pi/180)
plt.plot(x,y)
plt.show()
plt.savefig("filename.png",dpi=300,format="png")
```

####  給定散點圖顏色

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

plt.scatter(X,Y,c = 'r')
plt.show()
```
