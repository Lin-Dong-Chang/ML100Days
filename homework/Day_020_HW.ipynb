#### 繪製模型殘差 圖型


```python
import numpy as np
import seaborn as sns

x = np.random.normal(2, 1, 75)
y = 2 + 1.5 * x + np.random.normal(0, 2, 75)

sns.residplot(x,y)
```

#### 1. 使用 distplot()使用簡單的規則來正確猜測預設情況下正確的數位,但嘗試更多或更少的 bin 可能會顯示資料中的其他特徵
#### 2. 有無kde對圖形分布的影響

```python
import matplotlib.pyplot as plt
np.random.seed(0)
plt.figure(figsize=(20,5))
x = np.random.randint(500,size=1000)
plt.subplot(1,3,1)
sns.distplot(x, bins=20, kde = False)   #bins 分成幾個區間

plt.subplot(1,3,2)
sns.distplot(x, bins=10, kde = False)   #使用更小的bis但劃分較不明顯

plt.subplot(1,3,3)
sns.distplot(x, bins=20, kde = True)    #kde 核密度估計 
```
